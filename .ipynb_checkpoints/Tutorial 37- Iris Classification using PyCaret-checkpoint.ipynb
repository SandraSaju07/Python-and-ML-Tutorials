```python
# Import required libraries
from pycaret.datasets import get_data
from pycaret.classification import *
```

```python
# Load dataset
data = get_data('iris')
```

```python
# Set up environment
clf = setup(data, target='species')
```

```python
# Compare models to find the best model
compare_models()
```

```python
# Train model using K-Nearest Neighbors classifier
knn = create_model('knn')
```

```python
# Deploy the trained model
deploy_model(model=knn, model_name='knn_aws', platform='aws', authentication={'bucket':'pycaret-test'})
```

```python
# Load the saved model from AWS S3 bucket
knn_model = load_model('knn_aws', platform='aws', authentication={'bucket':'pycaret-test'})
```

```python
# Make predictions using loaded model
new_data = data.sample(5)

predictions = predict_model(knn_model, data=new_data)
predictions
```
