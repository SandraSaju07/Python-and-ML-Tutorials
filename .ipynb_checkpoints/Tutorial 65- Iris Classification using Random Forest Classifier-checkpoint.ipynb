```python
# Loading the library with Iris dataset
from sklearn.datasets import load_iris
```

```python
# Loading scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
```

```python
# Loading pandas
import pandas as pd
```

```python
# Loading numpy
import numpy as np
```

```python
# Setting random seed - keeps randomness reproducible
np.random.seed(0)
```

```python
# Creating an object called iris with the iris data
iris = load_iris()

iris
```

```python
# Creating a dataframe with the four feature variables
df = pd.DataFrame(iris.data, columns=[iris.feature_names])
```

```python
# Viewing the top 5 rows
df.head()
```

```python
# Adding a new column for the species name
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# Viewing the top 5 rows
df.head()
```

```python
df.columns
```

```python
# Flatten MultiIndex columns
df.columns = [' '.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

df.columns
```

```python
# Creating train and test data
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
```

```python
# View the top 5 rows
df.head()
```

```python
# Creating dataframes with train rows and test rows
train, test = df[df['is_train']==True], df[df['is_train'] == False]
```

```python
# Show the number of observations for train and test dataframes
print(f"No. of observations in train dataset: {len(train)}")
print(f"No. of observations in test dataset: {len(test)}")
```

```python
# Create a list of the feature column's names
features = df.columns[:4]

# View features
features
```

```python
# Converting each species name into digits
y = pd.factorize(train['species'])[0]
```

```python
# Viewing target
y
```

```python
# Creating a Random Forest classifier
clf = RandomForestClassifier(n_jobs=2, random_state=0)
```

```python
clf.fit(train[features], y)
```

```python
test[features].head()
```

```python
# Applying the trained classifier to the test data
clf.predict(test[features])
```

```python
test.head()
```

```python
# Viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[:10]
```

```python
# Mapping names for the plants for each predicted plant class
preds = iris.target_names[clf.predict(test[features])]
```

```python
# View the preidcted species for the first twenty five observations
preds[:25]
```

```python
# Viewing actual species for the first 25 test observations
test['species'].head(25)
```

```python
# Creating Confusion Matrix
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test['species'], preds)
```

```python
# Predicting new flowers
preds = iris.target_names[clf.predict([[6.0, 3.6, 5.4, 2.0], [5.0, 3.6, 1.4, 2.0]])]

preds
```
