```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('./data/canada_per_capita_income.csv')

data.head()
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(data.year,data['per capita income (US$)'],color='red',marker='+')
```

```python
model = linear_model.LinearRegression()
model.fit(data[['year']],data[['per capita income (US$)']])
```

```python
model.predict(np.array([2020]).reshape(-1,1))
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(data.year,data['per capita income (US$)'],color='red',marker='+')
plt.plot(data.year,model.predict(data[['year']]))
```
