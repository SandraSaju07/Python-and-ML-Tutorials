### Distribution Plots

* pairplot - more than 2 features

```python
import seaborn as sns
```

#### JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

#### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### PairPlot
A pair plot is also known as scatter plot, in which one variable in the same data row is matched with another variable's value, like this, showing all variables paired with all other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

#### DistPlot
Dist Plot helps us to find the distribution of columns - histogram

```python
sns.distplot(df['tip'])
```

```python
sns.displot(df['tip'])
```

```python
sns.histplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
sns.displot(df['tip'],kde=True,bins=10)  # kde means kernal density estimation
```
