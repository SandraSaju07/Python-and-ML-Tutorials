### Categorical Plots
Seaborn also helps us to do the analysis of categorical data points as well. This includes,
* boxplot
* violinplot
* countplot
* barplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df['sex'].unique()
```

```python
sns.countplot(x='sex',data=df)
```

```python
df['smoker'].unique()
```

```python
sns.countplot(x='smoker',data=df)
```

```python
df['day'].unique()
```

```python
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

#### Box Plot
A box and whisker plot also known as box plot is a graph that presents information from a five number summary.

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(y='total_bill',x='smoker',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(y='total_bill',data=df,hue='day')
```

```python
sns.boxplot(y='total_bill',hue='day',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=df)
```

#### Violin Plot
Violin Plot helps us to see the distribution of data in terms of both kernal density estimation and the box plot

```python
sns.violinplot(y='total_bill',x='day',data=df,palette='rainbow')
```
